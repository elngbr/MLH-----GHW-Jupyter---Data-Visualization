```python
# line
from matplotlib import pyplot as plt
# x values  
x= [5,2,9]
# y values
y=[10,8, 9]

# x, y -> 3,10



plt.plot(x, y)
plt.show()

```

```python
#line
from matplotlib import pyplot as alias
# x values

x=[5,8,8,8,8]
#y values

y=[8,8,8,8,8]

# x, y->5, 8

alias.plot(x, y)
alias.show()


```

```python
from matplotlib import pyplot as elena

a=[2,8, 4, 9]
b=[9,6, 4, 8]

elena.plot(a, b)
elena.show()
```

```python
#bar

elena.bar(a,b)
elena.show()
```

```python
elena.barh(a,b)
elena.show()
```

```python
elena.hist(a)
elena.show()
```

```python
elena.scatter(a,b)
elena.plot()
elena.show()
```

```python
elena.plot(a, b, 'o')
elena.show()
```

```python
elena.plot(a, b, marker="*")
elena.show()
```

```python
elena.plot(a, b, linestyle="dotted")
elena.show()
```

```python
students = ["Student1", "Student2", "Student3"]
# could have also used 
sizes=[35, 30, 35]
plt.pie(sizes, labels=students)
#                                           SEE                    YOU USED PLT, NOT YOUR ALIAS
plt.show()
```

```python
plt.pie(sizes, labels=students, autopct="%1.1f%%", startangle=90)
plt.title("Student Data")
plt.show()
```

```python
plt.pie(sizes, labels=students, autopct="%1.1f%%", startangle=70)
plt.title("Student Data")
plt.show()
```

```python
plt.pie(sizes, labels=students, autopct="%1.1f%%", startangle=120)
plt.title("Student Data")
plt.show()
```

```python
# explode

myExplode=[0.4, 0, 0]
plt.pie(sizes, labels=students, explode=myExplode, shadow=True)
plt.legend()
plt.show()
```
